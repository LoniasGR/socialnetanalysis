```python
# Ορίστε τον αριθμό κόμβων n=1x0 που θα χρησιμοποιήσετε για την κατασκευή των συνθετικών τοπολογιών.
# Όπου x αντικαταστήστε το τελευταίο ψηφίο του αριθμού μητρώου σας.

n = 120
```

```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Α) Δημιουργία και οπτικοποίηση σύνθετων τύπων δικτύου

Κατασκευή και οπτικοποίηση της τοπολογίας REG για βαθμό πλέγματος $d = 4$

```python
d = 4

# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. Επιλέξτε προσεκτικά τις παραμέτρους. 
REG = nx.watts_strogatz_graph(n,d,0) 

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RG(ER) για αριθμό συνδέσεων $M = 750$

```python
M = 750

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n,M)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RGG σε περιοχή $L\times L = 1$ (unit cube) και ακτίνα σύνδεσης $R=0.25$

```python
R = 0.25
L = 1
#n = 100

# για την κατασκευή του RGG, χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R) 

# για την οπτικοποίηση, χρησιμοποιήστε draw_network_edges, draw_network_nodes. 
# Για τη θέση των κόμβων στην τετράγωνη περιοχή, χρησιμοποιήστε την get_node_attributes από το networkx.
pos = nx.get_node_attributes(RGG, 'pos')
nx.draw_networkx_nodes(RGG, pos)
nx.draw_networkx_edges(RGG, pos)
```

Κατασκευή και οπτικοποίηση της τοπολογίας SF για βαθμό αρχικού πλέγματος $d=4$

```python
d = 4

# για την κατασκευή του SF, χρησιμοποιήστε την barabasi_albert_graph από το networkx.
SF = nx.barabasi_albert_graph(n, d) 

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SF)
```

Κατασκευή και οπτικοποίηση της τοπολογίας SW για βαθμό αρχικού πλέγματος $d=4$ και πιθανότητας ανασύνδεσης $p=0.3$

```python
d = 4
p = 0.3

# για την κατασκευή του SW, χρησιμοποιήστε την watts_strogatz_graph από το networkx.
SW = nx.watts_strogatz_graph(n, d, p)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SW)
```

### B) Μελέτη βαθμού κόμβων

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS)

##### 1)  Να υπολογιστεί ο βαθμός κάθε κόμβου

```python
# χρησιμοποιήστε την degree του networkx 

def find_degree(G):
    degrees = list()
    for n in G.nodes:
        degrees.append((n, G.degree(n)))
    return degrees
        

REG_degree = find_degree(REG)
RGER_degree = find_degree(RGER)
RGG_degree = find_degree(RGG)
SF_degree = find_degree(SF)
SW_degree = find_degree(SW)
```

##### 2)  Να αναπαρασταθεί (plot) η κατατανομή βαθμών κόμβων του δικτύου (δηλαδή, για κάθε βαθμό πόσοι κόμβοι έχουν αυτό το βαθμό) καθώς και η συγκεντρωτική κατανομή βαθμού κόμβου (δηλαδή για κάθε βαθμό πόσοι κόμβοι έχουν το πολύ αυτό το βαθμο)

```python
def plot_deg_stats(lst_deg):
    
    #για την κατανομή βαθμού κόμβου, χρησιμοποιήστε την hist του matplotlib
    values = [dgs[1] for dgs in lst_deg]
    plt.subplot(1, 2, 1)
    plt.hist(values)
    
    #για τη συγκεντρωτική κατανομή του διανύσματος values (σε αυτή την περίπτωση τa degree values),
    #χρησιμοποιήστε την plot(X,Y) του matplotlib
    n = len(lst_deg)
    X = np.sort(values) 
    Y = np.array(range(n))/float(n)
    plt.subplot(1, 2, 2)
    plt.plot(X,Y)
```

##### 3)  Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε μια από τις εν λόγω τοπολογίες

```python
def mean_var_degree(lst_deg):
    #χρησιμοποιήστε τις mean, var του numpy
    values = [dgs[1] for dgs in lst_deg]
    mean = np.mean(values)
    var = np.var(values)
    return (mean, var)
```

<b>Σχολιάστε εδώ τα αποτελέσματα των ερωτημάτων Β1, Β2, Β3.</b>

<b> Κανονικός Γράφος </b>

Ο κανονικός γράφος, όπως γνωρίζουμε έχει ίδιο αριθμό ακμών από κάθε κόμβο, αρα αναμέναμε ότι ο βαθμός θα είναι 4 παντού. Αυτό φαίνεται και από τα γραφήματα και από τον μέσο όρο και διασπορά.

```python
plot_deg_stats(REG_degree)
mean, var = mean_var_degree(REG_degree)

print("Mean is:", mean, " and variance is:", var)
```

Αυτό συμβαίνει γιατί κάθε ακμή ενώνει 2 γράφους.

<a name="RGG"><b> Τυχαίος Γεωμετρικός Γράφος (RGG) </b></a>

```python
plot_deg_stats(RGG_degree)
mean, var = mean_var_degree(RGG_degree)

print("Mean is:", mean, " and variance is:", var)
```

Ο τρόπος που δημιουργήται ο τυχαίος γεωμετρικός γράφος είναι τοποθετώντας τυχαία μέσα σε ένα μοναδιαίο κύβο κόμβους και μετα συνδέοντας αυτούς που απέχουν μέχρι 0.25 μεταξύ τους. Αρχικά απο θεωρία γνωρίζουμε ότι ο αναμενόμενος αριθμός γειτόνων για κάθε γράφο είναι $\frac{\pi r^2}{L^2}n$ που στην περίπτωση μας είναι $\pi * 0.25^2 * 120 = 23.6$. ### Γ) Δίκτυα με βάρη

Για ένα δίκτυο με βάρη στις συνδέσεις του, όμοια με τον πίνακα γειτνίασης, ορίζεται ο πίνακας βαρών, $W=[w_{ij}]$, όπου με $w_{ij}>0$ συμβολίζεται το βάρος στην ακμή που ενώνει τους κόμβους $i,j$. Για μη-κατευθυνόμενους γράφους ο πίνακας $W$ είναι συμμετρικός.

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να δημιουργηθεί τυχαίος πίνακας βαρών. Το εύρος των βαρών να κυμαίνεται στο διάστημα $[1,10]$.

```python
def add_weights(G):
    # για τιμές στο [1,10]: α=1, b=10, weight=np.random.rand()*(b-a) + a
    # για την προσθήκη βαρών στις ακμές, χρησιμοποήστε την add.edge του networkx
    
    a = 1
    b = 10
    new_G = nx.Graph()
    new_G.add_nodes_from(G.nodes())
    
    for u, v in G.edges():
        weight = np.random.rand()*(b-a) + a
        new_G.add_edge(u, v, weight=weight)
    return new_G
```

```python
REG_weighted = add_weights(REG)
RGER_weighted = add_weights(RGER)
RGG_weighted = add_weights(RGG)
SF_weighted = add_weights(SF)
SW_weighted = add_weights(SW)

print(len(REG_weighted.adj[0].items()))
```

##### i) Να υπολογιστεί η δύναμη του κάθε κόμβου

```python
#δίνεται η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α:
def strength(topology):
    weights_list=[]
    for i in range(0,len(topology.adj)):
        a=[]
        for j in range (0,len(topology.adj[i].items())):
            d=list(topology.adj[i].values())[j]
            c=list(d.values())[0]
            a.append(c)
        weights_list.append(a)    
    node_weights = []
    for node in weights_list:
        node_weights.append(sum(node))
    return node_weights

REG_strength = strength(REG_weighted)
RGER_strength = strength(RGER_weighted)
RGG_strength = strength(RGG_weighted)
SF_strength = strength(SF_weighted)
SW_strength = strength(SW_weighted)
```

##### ii) Να αναπαρασταθεί η συγκεντρωτική κατανομή δύναμης (cumulative strength distribution) των κόμβων του δικτύου (σε κάθε τιμή δύναμης αντιστοιχίζεται το πλήθος των κόμβων με δύναμη μικρότερη ή ίση από αυτή)

```python
def plot_cum_str_dis(stren):
    n = len(stren)
    X = np.sort(stren) 
    Y = np.array(range(n))/float(n)
    plt.plot(X,Y)
```

##### iii) Να υπολογιστεί η μέση δύναμη για όλους τους κόμβους

```python
def mean_strength(stren):
    #χρησιμοποιήστε τη mean του numpy
    mean = np.mean(stren)
    return mean
```

Σύμφωνα με το paper "The architecture of complex weighted netowrks", η έννοια της δύναμης ενός γράφου δημιουργήθηκε ως αντιστοιχία με το βαθμό στούς γράφους χωρίς βάρη. Μέσω αυτού μπορούμε να μετρήσουμε πόσο σημαντικός είναι ένας κόμβος του δικτύου μας. Η δύναμη υπολογίζεται ως το άθροισμα των βαρών των ακμών του γράφου που συνδέονται με τον συγκεκριμένο κόμβο.

### Δ) Υπολογισμός μέσου μήκους μονοπατιού

Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS) του πίνακα, να υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του στο δίκτυο. Να παρουσιαστούν συγκεντρωτικά τα αποτελέσματα. Πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως εξηγήθηκαν στη θεωρία.

```python
def compute_path_lengths(G):
    # χρησιμοποήστε την all_pairs_shortest_path_length του networkx 
    length = list(nx.all_pairs_shortest_path_length(G))
    dictlist = list()
    for item in length:
        temp = list(item[1].values())
        dictlist.append(temp)
    return dictlist

REG_path_lengths = compute_path_lengths(REG)
RGER_path_lengths = compute_path_lengths(RGER)
RGG_path_lengths = compute_path_lengths(RGG)
SF_path_lengths = compute_path_lengths(SF)
SW_path_lengths = compute_path_lengths(SW)

print("REGULAR GRAPH")
print("Mean path length REG: %f" %np.mean(REG_path_lengths))
print("Variance of path length REG: %f" %np.var(REG_path_lengths))
print("---------------------------------------------------------------")

print("RANDOM GRAPH - ERDOS-RENYI")
print("Mean path length RGER: %f" %np.mean(RGER_path_lengths))
print("Variance of path length RGER: %f" %np.var(RGER_path_lengths))
print("---------------------------------------------------------------")

print("RANDOM GEOMETRIC GRAPH")
print("Mean path length RGG: %f" %np.mean(RGG_path_lengths))
print("Variance of path length RGG: %f" %np.var(RGG_path_lengths))
print("---------------------------------------------------------------")

print("SCALE FREE GRAPH")
print("Mean path length SF: %f" %np.mean(SF_path_lengths))
print("Variance of path length SF: %f" %np.var(SF_path_lengths))
print("---------------------------------------------------------------")

print("SMALL WORLD GRAPH")
print("Mean path length SW: %f" %np.mean(SW_path_lengths))
print("Variance of path length SW: %f" %np.var(SW_path_lengths))
```

Το μέσο μήκος μονοπατιού υπολογίζεται από τον τύπο:

$$   \ell = \frac{1}{N(N-1)} \sum_{i \ne j} d_{ij} $$

όπου Ν: πλήθος κόμβων, $d_{ij}$ η ελάχιστη απόσταση μεταξύ δύο κόμβων.

Για τον **Regular Graph**, βλέπουμε ότι το μέσο μήκος μονοπατιού είναι αρκετά μικρό. Αυτό συμβαίνει γιατί υπάρχουν πολλές διαδρομές μικρού μήκους στον γράφο, αλλά και πολλές μεγάλου μήκους, με τις μεγαλύτερες να είναι μήκους $\frac{n}{4} = 30$. Αυτό φαίνεται και από την διασπορά, η οποία είναι πολύ μεγάλη. 

Για τον **Random Graph - Erdos-Renyi**, επειδή έχουμε πολλές ακμές (750) σε σχέση με τους κόμβους (120), περιμέναμε ότι οι περισσότεροι κόμβοι θα είναι σε αρκετά μικρή απόσταση, κάτι που φαίνεται να επιβεβαιώνεται από το μέσο μήκος των μονοπατιών και την διασπορά τους.

```python
# χρησιμοποιήστε την clustering του networkx για τον υπολογισμό του ΣΟ

def calculate_clust_coef(G):
    return list(nx.clustering(G).values())

def plot_cum_coef(coef):
    n = len(coef)
    X = np.sort(coef) 
    Y = np.array(range(n))/float(n)
    plt.plot(X,Y)

REG_clust_coef = calculate_clust_coef(REG)
RGER_clust_coef = calculate_clust_coef(RGER)
RGG_clust_coef = calculate_clust_coef(RGG)
SF_clust_coef = calculate_clust_coef(SF)
SW_clust_coef = calculate_clust_coef(SW)

#συγκεντρωτική κατανομή του ΣΟ
plt.figure()
plot_cum_coef(REG_clust_coef)
plt.figure()
plot_cum_coef(RGER_clust_coef)
plt.figure()
plot_cum_coef(RGG_clust_coef)
plt.figure()
plot_cum_coef(SF_clust_coef)
plt.figure()
plot_cum_coef(SW_clust_coef)
```

<h4>Αναλυτικός υπολογισμός κεντρικοτήτων</h4>

**Degree Centrality**: 
Το Degree Centrality για κάθε κόμβο υπολογίζεται ώς ο βαθμός του. To Freeman centrality υπολογίζεται ως:
$$ C_D(G)= \frac{\sum^{|V|}_{i=1} [C_D(v*)-C_D(v_i)] }{|V|^2-3|V|+2}$$
όπου $C_D(v*)$ είναι το μεγαλύτερο degree centrality στον γράφο.

**Closeness Centrality**:
To Closeness Centrality για κάθε κόμβο υπολογίζεται ως το ανίστροφο του άθροισματος των ελάχιστων μονοπατιών από κάθε άλλο κόμβο, δηλαδή $$ C_P(k) = \frac{1}{\sum^{n}_{j=1}{d(A, j)}}$$. Όταν θέλουμε να είναι κανονικοποιημένο, ώστε να μπορούμε να συγκρίνουμε μεταξύ γράφων, το πολλαπλασιάζουμε με **n-1**, όπου n είναι το πλήθος των κόμβων, δηλαδή: $$C_P(k) = \frac{n-1}{\sum_{j=1}^{n}{d(A, j)}}$$

**Betweenness Centrality**:
Το Betweeness Centrality για κάθε κόμβο x υπολογίζεται ως το άθροισμα των ελάχιστων μονοπατιών μεταξύ κάθε άλλων γράφων που περνάνε από τον x διαιρεμένο με το συνολικό πλήθως των ελάχιστων μονοπατιών μεταξύ των άλλων δυο γράφων, δηλαδη: $$C_B(v)= \sum_{s \neq v \neq t \in V}\frac{\sigma_{st}(v)}{\sigma_{st}}$$.

όπου $\sigma _{st}$ είναι ο συνολικός αριθμός ελάχιστων μονοπατιών μεταξύ του κόμβου $s$ και του $t$ και $\sigma_{st}(v)$ είναι ο αριθμώς των ελαχιστών μονοπατιών που περνάνε από τον $v$.

<p>Τοπολογία Α </p>

**Degree Centrality**:  Ο κόμβος 1 έχει βαθμό 3, ο κόμβος 2 έχει βαθμό 1, ο κόμβος 3 έχει βαθμό 1 και ο κόμβος 4 έχει βαθμό 1. Άρα το Freeman centrality είναι: $\frac{(3-3) + (3-1) + (3-1) + (3-1)}{3*2} = \frac{6}{6} = 1$. </br>
**Closeness Centrality**: Ο κόμβος 1 έχει μήκος ελάχιστου μονοπατιού 1 για άλλο κόμβο, αρα το closeness centrality είναι $\frac{3}{1+1+1} = 1$. Oι κόμβοι 2, 3 και 4 έχουν απόσταση 1 από τον κόμβο 1 και απόσταση 2 από όλους τους άλλους, αρα και οι τρείς έχουν το ίδιο centrality που είναι: $\frac{3}{1 + 2 + 2} = \frac{3}{5} = 0.6$. </br>
**Betweenness Centrality**: Υπάρχει ένα ελάχιστο μονοπάτι μεταξύ των κόμβων $\{2, 3, 4\}$ ανα δύο που περνάει μέσα από τον 1 και αυτά είναι όλα τα ελάχιστα μονοπάτια, αρα θα έχουμε: $\frac{1}{1} + \frac{1}{1} + \frac{1}{1} = 3$. Για τους υπόλοιπους κόμβους δεν υπάρχει κανένα ελάχιστο μονοπάτι μεταξύ των άλλων κόμβων που να περνάει μέσα από αυτούς, αρα το Betweeness Centrality τους θα είναι 0.

<p>Τοπολογία B </p>

**Degree Centrality**: Ο κόμβος 1 έχει βαθμό 3, ο κόμβος 2 έχει βαθμό 3, ο κόμβος 3 έχει βαθμό 3 και ο κόμβος 4 έχει βαθμό 3. Aρα το Freeman centrality είναι: $\frac{(3-3) + (3-3) + (3-3) + (3-3)}{3*3} = 0$. </br>
**Closeness Centrality**: Όλοι οι κόμβοι απέχουν απόσταση 1 μεταξύ τους, αρα όλων το closeness centrality θα είναι: $\frac{3}{1+1+1} = 1$. </br>
**Betweenness Centrality**: Σε αυτή την τοπολογία κανένας κόμβος δεν είναι μέρος ενός ελάχιστου μονοπατιού μεταξύ κάποιων άλλων, άλλα για όλους τους κόμβους το Betweeness Centrality θα είναι 0.


Συμπεράσματα:

**Degree Centrality**:  Ο γράφος 1 έχει μεγαλύτερο *degree centrality* από τον γράφο 2, γεγονός φανερό, αφου στην τοπολογία 1 ο κόμβος 1 λειτουργεί σαν συνδετικός κόμβος για όλους τους υπόλοιπος.</br>
**Closeness Centrality**: Είναι εύκολο να δούμε ότι η μέση τιμή του *closeness centrality* είναι αρκετά μεγαλύτερη για την τοπολογία 2, αφού είναι πλήρως συνδεδεμένη.</br>
**Betweenness Centrality**: Παρατηρούμε ότι σε ένα πλήρος συνδεδεμένο γράφο, το *betweeness centrality* θα είναι 0. Αντίθετα στην τοπολογία 1, ο κόμβος 1 προφανώς λειτουργεί σαν συνδετικός κρίκος μεταξύ των υπόλοιπων κόμβων.

```python
def calculate_degree_centrality(G):
    degree_centrality = nx.degree_centrality(G)
    return list(degree_centrality.values())

def calculate_closeness_centrality(G):
    closeness_centrality = nx.closeness_centrality(G)
    return list(closeness_centrality.values())

def calculate_betweenness_centrality(G):
    betweenness_centrality = nx.betweenness_centrality(G)
    return list(betweenness_centrality.values())

def calculate_eigenvector_centrality(G):
    eigenvector_centrality = nx.eigenvector_centrality(G)
    return list(eigenvector_centrality.values())

REG_degree_centrality = calculate_degree_centrality(REG)
REG_closeness_centrality = calculate_closeness_centrality(REG)
REG_betweenness_centrality = calculate_betweenness_centrality(REG)
REG_eigenvector_centrality = calculate_eigenvector_centrality(REG)

RGER_degree_centrality = calculate_degree_centrality(RGER)
RGER_closeness_centrality = calculate_closeness_centrality(RGER)
RGER_betweenness_centrality = calculate_betweenness_centrality(RGER)
RGER_eigenvector_centrality = calculate_eigenvector_centrality(RGER)

RGG_degree_centrality = calculate_degree_centrality(RGG)
RGG_closeness_centrality = calculate_closeness_centrality(RGG)
RGG_betweenness_centrality = calculate_betweenness_centrality(RGG)
RGG_eigenvector_centrality = calculate_eigenvector_centrality(RGG)

SF_degree_centrality = calculate_degree_centrality(SF)
SF_closeness_centrality = calculate_closeness_centrality(SF)
SF_betweenness_centrality = calculate_betweenness_centrality(SF)
SF_eigenvector_centrality = calculate_eigenvector_centrality(SF)

SW_degree_centrality = calculate_degree_centrality(SW)
SW_closeness_centrality = calculate_closeness_centrality(SW)
SW_betweenness_centrality = calculate_betweenness_centrality(SW)
SW_eigenvector_centrality = calculate_eigenvector_centrality(SW)

# για πιο εύκολο σχολιασμό, να παράξετε 1 figure για όλες τις τοπολογίες ανά κεντρικότητα
def plot_centrality(centrality):
    graphs = list()
    labels = ['REG', 'RGER', 'RGG', 'SF', 'SW']
    lines = ['k--', 'b:', 'r-.', 'y-', 'm--']
    if centrality == 'degree':
        graphs.append(REG_degree_centrality)
        graphs.append(RGER_degree_centrality)
        graphs.append(RGG_degree_centrality)
        graphs.append(SF_degree_centrality)
        graphs.append(SW_degree_centrality)
    elif centrality == 'closeness':
        graphs.append(REG_closeness_centrality)
        graphs.append(RGER_closeness_centrality)
        graphs.append(RGG_closeness_centrality)
        graphs.append(SF_closeness_centrality)
        graphs.append(SW_closeness_centrality)
    elif centrality == 'betweenness':
        graphs.append(REG_betweenness_centrality)
        graphs.append(RGER_betweenness_centrality)
        graphs.append(RGG_betweenness_centrality)
        graphs.append(SF_betweenness_centrality)
        graphs.append(SW_betweenness_centrality)
    elif centrality == 'eigenvector':
        graphs.append(REG_eigenvector_centrality)
        graphs.append(RGER_eigenvector_centrality)
        graphs.append(RGG_eigenvector_centrality)
        graphs.append(SF_eigenvector_centrality)
        graphs.append(SW_eigenvector_centrality)
    else:
        return -1
    for i in range(len(graphs)):
        graph = graphs[i]
        lab = labels[i]
        line = lines[i]
        n = len(graph)
        X = np.sort(graph) 
        Y = np.array(range(n))/float(n)
        plt.plot(X,Y, line, label=lab)
    plt.legend()
    plt.title(centrality)
```

```python
plt.figure()
plot_centrality('degree')
plt.figure()
plot_centrality('closeness')
plt.figure()
plot_centrality('betweenness')
plt.figure()
plot_centrality('eigenvector')
```

**Degree Centrality**: Παρατηρούμε ότι ιδιαίτερο ενδιαφέρον σε αυτή τηνγ γραφική έχει ο *Scale Free* γράφος, καθώς οι περισσότεροι κόμβοι του έχουν μικρό Degree Centrality, εκτός από λίγους που έχουν αρκετά μεγάλο. Αυτό το γεγονός επαληθεύει τα θεωρητικά αποτελέσματα που έχουμε για τον Scale Free γράφο. Ακόμα, ο *RGG* έχει σχετικά μεγάλο Degree Centrality για τους λόγους που αναφέραμε και νωρίτερα. Οι *REG* και *SW*, όπως περιμέναμε έχουν μικρό Degree Centrality για όλους τους κόμβους τους και ο *RGER* έχει ελαφρώς μεγαλύτερο, λόγω του μεγάλου πλήθους ακμών που επιλέξαμε.

**Closeness Centrality**: Εδώ όπως περιμέναμε ο *SF* έχει το μεγαλύτερο closeness centrality για λίγους από τους κόμβους του, για λόγους που εξηγήσαμε και νωρίτερα στα ελάχιστα μονοπάτια. Ενδιαφέρον έχει ότι ο *RGER* έχει επίσης αρκετά μεγάλο centrality, γεγονός που ήταν αναμενόμενο, αφού νωρίτερα είχαμε δει ότι γενικά έχει μικρό μέσο μήκος μονοπατιών. Αντίθετα ο *SW* έχει σχετικά αρκετά μεγάλο μέσο μήκος μονοπατιών και φαίνεται και εδώ.

**Betweenness Centrality**: To Betweenness centrality είναι αρκετά μεγάλο στον *REG*, καθώς κάθε κόμβος του αποτελεί μέρος ελάχιστου μονοπατιού για κάποιους άλλους κόμβους. Ακόμα για τον *SF* παρατηρούμε ότι υπάρχουν λίγοι κόμβοι που λειτουργούν σαν μέρος ελάχιστου μονοπατιού για πολλά μονοπάτια. Ξανά αυτό είναι κάτι αναμενόμενο από την θεωρία. Ομοίως αλλά λιγότερο για τον *SW*. Τα χειρότερα αποτελέσματα είναι των *RGER* και *RGG* καθώς η κατανιμή των ελάχιστων μονοπατιών είναι τυχαία και έτσι δεν υπάρχει κάποιος πιο κεντρικός κόμβος.

**Eigenvector Centrality**: Το Eigenvector Centrality δεν μας δίνει κάποιο ξεκάθαρο αποτέλεσμα. Τo centrality αυτό βασίζεται στην κεντρικότητα των γειτόνων για να υπολογίσει την κεντρικότητα κάποιου κόμβου. Όπως βλέπουμε μεγαλύτερο centrality έχει ξανά το *SF*, για λίγους κόμβους, ενώ τα υπόλοιπα (εκτός του *REG* που είναι σταθερό, αφού όλοι οι κόμβοι έχουν την ίδια μορφή) έχουν πολύ παρόμοια συγκεντρωτική γραφική.

Για κάθε έναν τύπο δικτύου, να παράξετε 100 διαφορετικές τοπολογίες και να παρουσιάσετε το διάγραμμα ποσοστού συνεκτικότητας σε σχέση με τις καθορισμένες παραμέτρους κάθε δικτύου.

```python
#για τον έλεγχο της συνεκτικότητας, χρησιμοποίηστε την is_connected του networkx
#για τη γραφική αναπαράσταση των αποτελεσμάτων, μπορείτε να χρησιμοποίησετε την bar του matplotlib
#έλεγχος συνεκτικότητας για REG

def check_connected_REG(n):
    connected = {i :0 for i in list(range(2,12,2))}
    for d in range(2, 12, 2):
        for i in range(100):
            REG = nx.watts_strogatz_graph(n,d,0) 
            if nx.is_connected(REG):
                connected[d] = connected[d] + 1

    y = list(connected.values())
    y = [v/100 for v in y]
    return y

#plot example:
def plot_connected_REG(n, y):
    plt.bar(list(range(2,12,2)),y, width=0.5, color='b')
    plt.title("REG n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Degree")
    plt.show()

plot_connected_REG(100, check_connected_REG(100))
plot_connected_REG(200, check_connected_REG(200))
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το REG:

Όπως βλέπουμε, ο κανονικός γράφος είναι πάντα συνδεδεμένος. Αυτό συμβαίνει από τον ορισμό του γράφου, καθώς κάθε κόμβος συνδέεται με τους d γείτονες του.

```python
#έλεγχος συνεκτικότητας για RGER
def check_connected_RGER(n):
    connected ={i :0 for i in list(range(100,900,100))}
    for M in range(100, 900, 100):
        for i in range(100):
            RGER = nx.gnm_random_graph(n,M)
            if nx.is_connected(RGER):
                connected[M] = connected[M] + 1

    y = list(connected.values())
    y = [v/100 for v in y]
    return y

#plot example:
def plot_connected_RGER(n, y):
    plt.bar(list(range(100,900,100)),y, width=5, color='b')
    plt.title("RGER n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Connections")
    plt.show()

plot_connected_RGER(100, check_connected_RGER(100))
plot_connected_RGER(200, check_connected_RGER(200)) 
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το RGER:

Όπως βλέπουμε για μικρά Μ, η συνεκτικότητα του RGER δεν είναι βέβαιη, αυτό συμβαίνει γιατί, για $Μ \leq n$, είναι αδύνατο ο γράφος να είναι συνδεδεμένος, αφού οι ακμές είναι λιγότερες από τις κορυφές. καθώς ο αριθμός των ακμών μεγαλώνει, ο γράφος είναι όλο και πιο πιθανό να είναι συνδεδεμένος. Με >500 ακμές και 100 κόμβους ο γράφος είναι πάντα συνδεδεμένος, ενώ για 200 κόμβους δεν είναι βέβαιο ακόμαι και με 800 ακμές αν θα είναι συνδεδεμένος πάντα.

```python
#έλεγχος συνεκτικότητας RG Gilbert
#για την κατασκευή της RG χρησιμοποιήστε τη συνάρτηση gnp_random_graph του networkx.
def check_connected_RG(n):
    connected ={i :0 for i in list(np.arange(0.1, 1.0, 0.1))}
    for p in np.arange(0.1, 1, 0.1):
        for i in range(100):
            RG = nx.gnp_random_graph(n,p)
            if nx.is_connected(RG):
                connected[p] = connected[p] + 1
    y = list(connected.values())
    y = [v/100 for v in y]
    return y

#plot example:
def plot_connected_RG(n, y):
    plt.bar(list(np.arange(0.1,1,0.1)),y, width=0.01, color='b')
    plt.title("RG n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Probability")
    plt.show()

plot_connected_RG(100, check_connected_RG(100))
plot_connected_RG(200, check_connected_RG(200)) 
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το RG Gilbert:

Σύμφωνα με την θεωρία, ο αναμενόμενος αριθμός ακμών για αυτόν τον γράφο είναι $$\binom{n}{2}p = \frac{n!}{2!(n-2)!}p = \frac{n(n-1)}{2}p$$.
Aν $n = 100$ τότε οι αναμενόμενες ακμές είναι $50 * 99 * p = 4950 * p$. Αρα ακόμα και για μικρα p (0.1), οι αναμενόμενες ακμές είναι πολύ περισσότερες από τους κόμβους, αρα αναμένουμε ο γράφος να είναι συνδεδεμένος, όπως και συμβαίνει. 

```python
#έλεγχος συνεκτικότητας RGG
def check_connected_RGG(n):
    connected ={i :0 for i in list(np.arange(0.025, 0.275, 0.025))}
    for R in np.arange(0.025, 0.275, 0.025):
        for i in range(100):
            RGG = nx.random_geometric_graph(n, R) 
            if nx.is_connected(RGG):
                connected[R] = connected[R] + 1
    y = list(connected.values())
    y = [v/100 for v in y]
    return y

#plot example:
def plot_connected_RGG(n, y):
    plt.bar(list(np.arange(0.025, 0.275, 0.025)),y, width=0.01, color='b')
    plt.title("RGG n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Radius")
    plt.show()

plot_connected_RGG(100, check_connected_RGG(100))
plot_connected_RGG(200, check_connected_RGG(200))
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το RGG:

Απο θεωρία, η ποσότητα $$\mu = ne^{-\pi r^{2}n}$$
μας λέει πόσο συνδεδεμένος θα είναι ο γράφος. Για $\mu \longrightarrow \infty$ ο γράφος είναι ασυμπτωτικά σχεδόν σίγουρα μη συνδεδεμένος, ενώ αν $\mu \longrightarrow 0$ θα είναι σχεδόν σίγουρα συνδεδεμένος. Στην περίπτωση μας για $r=0.025$ και $n=100$, έχουμε τιμή γύρω στο $80 \longrightarrow 0$. 

Για επαλήθευση, η πιθανότητα ένας κόμβος να είναι αποσυνδεδεμένος είναι $$(1-\pi r^{2})^{n-1}$$

που για τις παραπάνω τιμές είναι 0.86. Αυτό μας οδηγεί στο συμπέρασμα ότι για μικρές τιμές του $r$ ο γράφος θα είναι αποσυνδεδεμένος.

```python
def check_connected_SF(n):
    connected ={i :0 for i in list(range(2, 12, 2))}
    for d in np.arange(2, 12, 2):
        for i in range(100):
            SF = nx.barabasi_albert_graph(n, d) 
            if nx.is_connected(SF):
                connected[d] = connected[d] + 1
    y = list(connected.values())
    y = [v/100 for v in y]
    return y

#plot example:
def plot_connected_SF(n, y):
    plt.bar(list(range(2, 12, 2)), y, width=0.5, color='b')
    plt.title("SF n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("Degree")
    plt.show()


plot_connected_SF(100, check_connected_SF(100))
plot_connected_SF(200, check_connected_SF(200))
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το SF:

Όλοι οι SF γράφοι που δημιουργήθηκαν ήταν συνδεδεμένοι. Αυτό συμβαίνει γιατί το δίκτυο μας ξεκινά με m κόμβους (2-10) στην περίπτωση μας και σε κάθε εποχή προσθέτει έναν ακόμα κόμβο με 10 ακμές, άρα ακόμα και από την πρώτη εποχή ο γράφος θα είναι πάντα συνδεδεμένος.

```python
# για την κατασκευή της SW, χρησιμοποιήστε την connected_watts_strogatz_graph()
#𝑑∈[2,10]  με βήμα  2  και  𝑔𝑝∈[0.1,0.7]  με βήμα  0.1
def check_connected_SW(n):
    connected = dict()
    for i in list(range(2, 12, 2)):
        for j in list(np.arange(0.1, 0.8, 0.1)):
            connected.update({(i, j):0})
    for k in range(2, 12, 2):
        for p in np.arange(0.1, 0.8, 0.1):
            for i in range(100):
                SW = nx.connected_watts_strogatz_graph(n, k, p) 
                if nx.is_connected(SW):
                    connected[(k,p)] = connected[(k,p)] + 1
    y = list(connected.values())
    y = [v/100 for v in y]
    return y

#plot example:
def plot_connected_SW(n, y):
    x = list()
    for i in list(range(2, 12, 2)):
        for j in list(np.arange(0.1, 0.8, 0.1)):
            x.append('(' + str(i) + ', ' + str(round(j, 1)) + ')')
    plt.bar(x, y, width=0.5, color='b')
    plt.title("SW n = %d" %n)
    plt.ylabel("Number of connected graphs")
    plt.xlabel("(Neighbours, Probability)")
    plt.xticks(rotation=90, ha='right')
    plt.show()


plot_connected_SW(100, check_connected_SW(100))
plot_connected_SW(200, check_connected_SW(200))
```

Σχολιάστε τα αποτελέσματα συνεκτικότητας για το SW:

Εδώ αρχικά όλοι οι κόμβοι είναι συνδεδεμένοι με τους 2-10 κοντινότερους γείτονες τους σε τοπολογία δακτύλιου. Μετά με πιθανότητα 0.1-0.7 κάποιες από τις ακμές γίνονται rewire σε μακρινούς κόμβους. Όπως βλέπουμε, με αυτό τον τρόπο, ο συνολικός γράφος, πάντα μένει συνδεδεμένος.

- Σε ποιες περιπτώσεις εμφανίζονται φαινόμενα κατωφλίου;

Τα **φαινόμενα κατωφλίου** εμφανίζονται όταν πιθανότητα μιας ιδιότητας ενός τυχαίου γράφου μεταβαίνει από 0 σε 1 με μια μικρή αλλαγή στις παραμέτρους του.
Φαινόμενα κατοφλίου παρατηρούμε στους γράφους RGER για 100 κόμβους όπου από 200 ακμές και πιθανότητα 0.3 μεταβαίνουμε σε 300 ακμές και πιθανότητα 0.8. Επίσης εμφανίζεται στον RGG για μετάβαση από ακτίνα 0.25 σε 0.275, όπου η πιθανότητα πηδάει από 0.25 σε 0.8 ο γράφος να είναι συνδεδεμένος (100 κόμβοι). Αντίστοιχα για 200 κόμβους αυτό συμβαίνει από το 0.1 στο 0.125.

- Ποιες είναι οι κρίσιμες τιμές για κάθε τύπο δικτύου οι οποίες οδηγούν σε μετάβαση φάσης (phase transitions);

Οι τιμές αυτές είναι οι τιμές που αναφέραμε παραπάνω - (200 και 300 για RGER 100 κόμβων), (0.25 και 0.275 για RGG 100 κόμβων) και (0.1 και 0.125 για RGG 200 κόμβων). Τέλος υπάρχει μια πολύ πιο ομαλή μετάβαση στον RGER 200 κόμβων από 400 σε 700 ακμές. 

- Ποιες μεταβάσεις φάσης είναι απότομες (sharp) και ποιες ομαλές (smooth);

Οι τρεις πρώτες μεταβάσεις που αναφέρθηκαν ήταν απότομες, ενώ η τελευταία ήταν ομαλή.

### Κ) Εγω-κεντρικότητες (Ego-centralities)
Για τις συνθετικές τοπολογίες του πίνακα 1, να υπολογιστεί η ενδιαμεσική εγω-κεντρικότητα (ego-betweenness centrality) του κάθε κόμβου. Στη συνέχεια να καταταχθούν οι κόμβοι με βάση την ενδιαμεσική εγωκεντρικότητα και την ενδιαμεσική κεντρικότητα και να συγκριθούν/σχολιαστούν οι κατατάξεις ως προς την ομοιότητά τους για κάθε μία τοπολογία.  

```python
#για την κατασκευή των εγωδικτύων, χρησιμοποιήστε την ego_graph του networkx
#για τον υπολογισμό της ενδιαμεσικής εγωκεντρικότητας θα χρειαστείτε τις adjacency_matrix (networkx), to.dense, ones (numpy)
#linalg.matrix_power, multiply, triu, nonzero του numpy
#κατασκευάστε dataframe από το pandas με γραμμές τους κόμβους και στήλες τις τιμές των υπό εξέταση κεντρικοτήτων
#κατασκευάστε στη συνέχεια pairplot από το seaborn για να εξετάσετε τις συσχετίσεις μεταξύ τους.
def find_ego_centrality(G):
    ego_centrality = list()
    for n in G.nodes():
        ego_G = nx.ego_graph(G, n)
        A = nx.adjacency_matrix(ego_G).todense()
        geodesics = np.triu(np.multiply(np.linalg.matrix_power(A, 2), (np.ones(A.shape) - A)))
        entries = geodesics[np.nonzero(geodesics)]
        sum = 0
        for e in entries:
            sum = sum + 1/e
        ego_centrality.append(sum)
    return ego_centrality
```

```python
n = 120
d = 4
M = 750
R = 0.25
p = 0.3

REG = nx.watts_strogatz_graph(n,d,0)
RGER = nx.gnm_random_graph(n,M)
RGG = nx.random_geometric_graph(n, R) 
SF = nx.barabasi_albert_graph(n, d) 
SW = nx.watts_strogatz_graph(n, d, p)

REG_ego_centrality = find_ego_centrality(REG)
REG_betweenness_centrality = calculate_betweenness_centrality(REG)

RGER_ego_centrality = find_ego_centrality(RGER)
RGER_betweenness_centrality = calculate_betweenness_centrality(RGER)

RGG_ego_centrality = find_ego_centrality(RGG)
RGG_betweenness_centrality = calculate_betweenness_centrality(RGG)

SF_ego_centrality = find_ego_centrality(SF)
SF_betweenness_centrality = calculate_betweenness_centrality(SF)

SW_ego_centrality = find_ego_centrality(SW)
SW_betweenness_centrality = calculate_betweenness_centrality(SW)

print('REG ego centrality values:')
print([round(f,3) for f in REG_ego_centrality])
print('REG ego centrality indexes sorted:')
print(np.argsort([round(f,3) for f in REG_ego_centrality]))
print('REG betweeness values:')
print([round(f,3) for f in REG_betweenness_centrality])
print('REG betweenness indexes sorted:')
print(np.argsort([round(f,3) for f in REG_betweenness_centrality]))
print('--------------------------------------------------------------------------------------')

print('RGER ego centrality values:')
print([round(f,6) for f in RGER_ego_centrality])
print('RGER ego centrality indexes sorted:')
print(np.argsort([round(f,6) for f in RGER_ego_centrality]))
print('RGER betweeness values:')
print([round(f,6) for f in RGER_betweenness_centrality])
print('RGER betweenness indexes sorted:')
print(np.argsort([round(f,6) for f in RGER_betweenness_centrality]))
print('--------------------------------------------------------------------------------------')

print('RGG ego centrality values:')
print([round(f,6) for f in RGG_ego_centrality])
print('RGG ego centrality indexes sorted:')
print(np.argsort([round(f,6) for f in RGG_ego_centrality]))
print('RGG betweeness values:')
print([round(f,6) for f in RGG_betweenness_centrality])
print('RGG betweenness indexes sorted:')
print(np.argsort([round(f,6) for f in RGG_betweenness_centrality]))
print('--------------------------------------------------------------------------------------')

print('SF ego centrality values:')
print([round(f,6) for f in SF_ego_centrality])
print('SF ego centrality indexes sorted:')
print(np.argsort([round(f,6) for f in SF_ego_centrality]))
print('SF betweeness values:')
print([round(f,6) for f in SF_betweenness_centrality])
print('SF betweenness indexes sorted:')
print(np.argsort([round(f,6) for f in SF_betweenness_centrality]))
print('--------------------------------------------------------------------------------------')

print('SW ego centrality values:')
print([round(f,6) for f in SW_ego_centrality])
print('SW ego centrality indexes sorted:')
print(np.argsort([round(f,6) for f in SW_ego_centrality]))
print('SW betweeness values:')
print([round(f,6) for f in SW_betweenness_centrality])
print('SW betweenness indexes sorted:')
print(np.argsort([round(f,6) for f in SW_betweenness_centrality]))
```
